```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#read financial data  from a CSV file
finance_data = pd.read_excel('../data/processed/personal_finance_dataset.xlsx', sheet_name='Sheet1')
finance_data
```

```python
finance_data = finance_data[finance_data['Type'] == 'Expense']
finance_data
```

```python
group_category = finance_data.groupby('Category')['Amount'].sum().sort_values(ascending=False).reset_index()
group_category
```

```python
group_subcategory = finance_data.groupby('Subcategory')['Amount'].sum().sort_values(ascending=False).reset_index()
group_subcategory
```

```python
daily_spending = finance_data.groupby('Date')['Amount'].sum()
daily_average_spending = daily_spending.mean()
daily_average_spending
```
